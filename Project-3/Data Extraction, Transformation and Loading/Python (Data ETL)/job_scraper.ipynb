```python
# getting the number of links stored in mongo 
total_job_links = len(links_list) + 1

print('Beginning scrape')
for index, link in enumerate(links_list):

    # reset variables after every scrape
    job_title = None
    employer_name = None
    employer_location = None
    posted_time = None
    num_of_applicants = None
    job_description = None
    experience_level = None
    employment_type = None
    industry = None
    job_functions = None
    
#   open chrome browser and visit job page link
    driver.get(link['link'])
    print(f'Processing job {index + 1}/{total_job_links}: {link["link"]}')
    # clicking "see more" button on the job page to expose all the data
    try:
        element = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.XPATH,'//*[@class="t-14 t-black--light t-bold artdeco-button artdeco-button--muted artdeco-button--icon-right artdeco-button--2 artdeco-button--tertiary ember-view"]'))
        ).click()
        time.sleep(5)
    except:
        continue

    try:        
        # Job Title
          jobTitle = driver.find_element_by_xpath('//*[@class="jobs-top-card__job-title t-24"]').text
          
          if jobTitle is not None:
              job_title = driver.find_element_by_xpath('//*[@class="jobs-top-card__job-title t-24"]').text
          else:
              job_title = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/h1').text
    except:
        pass
    

    try: 
        # Employer Name
          employerName = driver.find_element_by_xpath('//*[@class="jobs-top-card__company-url ember-view"]').text
          
          if employerName is not None:
              employer_name = driver.find_element_by_xpath('//*[@class="jobs-top-card__company-url ember-view"]').text
          else:
              employer_name = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/h3/a[1]').text
    except:
        pass

    try: 
        # Employer Location
          employerLocation = driver.find_element_by_xpath('//*[@class="jobs-top-card__bullet"]').text
          
          if employerLocation is not None:
             employer_location = driver.find_element_by_xpath('//*[@class="jobs-top-card__bullet"]').text
          else:
             employer_location = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/h3/a[3]').text
    except:
        pass

    try:
           # Posted Time
          postedTime = driver.find_element_by_xpath('//*[@class="mt1 full-width flex-grow-1 t-14 t-black--light"]').text.split('\n')[1]
          
          if postedTime is not None:
              posted_time = driver.find_element_by_xpath('//*[@class="mt1 full-width flex-grow-1 t-14 t-black--light"]').text.split('\n')[1]
          else:
              posted_time = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/p').text.split('\n')[1]
    except:
        pass

    try:
          # Number of Applicants
          numApplicants = driver.find_element_by_xpath('//*[@class="mt1 full-width flex-grow-1 t-14 t-black--light"]').text
          
          if numApplicants is not None:
              num_of_applicants = driver.find_element_by_xpath('//*[@class="mt1 full-width flex-grow-1 t-14 t-black--light"]').text
              num_of_applicants = re.search(r'( \d+ applicants)',num_of_applicants).group(0)
              num_of_applicants = int(re.search(r'(\d+)',num_of_applicants).group(0))
          else:
              num_of_applicants = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[2]/div[1]/p').text
              num_of_applicants = re.search(r'( \d+ applicants)',num_of_applicants).group(0)
              num_of_applicants = int(re.search(r'(\d+)',num_of_applicants).group(0))       
    except:
        pass
    
    try: 
        # Job Description   
        jobDescription = driver.find_element_by_xpath('//*[@class="jobs-box__html-content jobs-description-content__text t-14 t-normal"]')
          
        if jobDescription is not None:
          job_description = driver.find_element_by_xpath('//*[@class="jobs-box__html-content jobs-description-content__text t-14 t-normal"]').text.lower() \
             .translate({ord(ch):' ' for ch in '. \n , - / \ [ ]( )'}).replace('  ', ' ')
        else:
          job_description = driver.find_element_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[3]/article/div/div[1]').text.lower() \
             .translate({ord(ch):' ' for ch in '. \n , - / \ [ ]( )'}).replace('  ', ' ')      
    except:
        print(f'Job {index+1} Description: does not exist?') 
        pass

    try: 
        # Experience Level
        experienceLevel = driver.find_elements_by_xpath('//*[@class="jobs-box__body js-formatted-exp-body"]')[0].text
        
        if experienceLevel is not None:
            experience_level = driver.find_elements_by_xpath('//*[@class="jobs-box__body js-formatted-exp-body"]')[0].text
        else:
            experience_level = driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[3]/article/div/div[2]/div/div[1]/p')[0].text
            
    except:
        pass

    try: 
        # Industry
        industryLinkedIn = driver.find_elements_by_xpath('//*[@class="jobs-box__list jobs-description-details__list js-formatted-industries-list"]')
        
        if industryLinkedIn is not None:
          industry1 = driver.find_elements_by_xpath('//*[@class="jobs-box__list jobs-description-details__list js-formatted-industries-list"]')
          industry2 = industry1[0].find_elements_by_tag_name('li')
          industry = [ind.text for ind in industry2]
        else:
          industry1 = driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[3]/article/div/div[2]/div/div[2]/ul')
          industry2 = industry1[0].find_elements_by_tag_name('li')
          industry = [ind.text for ind in industry2]
    except:
        print(f'Job {index+1} Industry: does not exist?')
        pass

    try: 
        # Employment Type
        employmentType = driver.find_elements_by_xpath('//*[@class="jobs-box__body js-formatted-employment-status-body"]')[0].text
        
        if employmentType is not None:
          employment_type = driver.find_elements_by_xpath('//*[@class="jobs-box__body js-formatted-employment-status-body"]')[0].text
        else:
          employment_type = driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[3]/article/div/div[2]/div/div[3]/p')[0].text
    except:
        print(f'Job {index+1} employment type: does not exist?')
        pass

    try: 
        # Job Functions
        jobFunction = driver.find_elements_by_xpath('//*[@class="jobs-box__list jobs-description-details__list js-formatted-job-functions-list"]')
        
        if jobFunction is not None:
          job_functions1 = driver.find_elements_by_xpath('//*[@class="jobs-box__list jobs-description-details__list js-formatted-job-functions-list"]')
          job_functions2 = job_functions1[0].find_elements_by_tag_name('li')
          job_functions = [jf.text for jf in job_functions2]
        else:
          job_functions1 = driver.find_elements_by_xpath('/html/body/div[8]/div[3]/div/div[1]/div[1]/div/div[1]/div/div[3]/article/div/div[2]/div/div[4]/ul')
          job_functions2 = job_functions1[0].find_elements_by_tag_name('li')
          job_functions = [jf.text for jf in job_functions2]

    except:
        print(f'Job {index+1} functions: does not exist?')
        pass
          
    linkedin_search_query = search_query.replace("%20"," ").replace("%26","&")

    jobs_collection_security.insert_one({
        'scrape_time': datetime.datetime.now(),
        'job_id': index,
        'job_title':job_title,
        'employer_name': employer_name,
        'employer_location':employer_location,
        'posted_time':posted_time,
        'num_of_applicants':num_of_applicants,
        'job_description': job_description,
        'experience_level': experience_level,
        'industry': linkedin_search_query,
        'LinkedIn industry': industry,
        'employment_type': employment_type,
        'job_functions':job_functions,
        'url_link' : link['link']
    })

print("Scrape Complete")
```
